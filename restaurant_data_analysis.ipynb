# 🍽️ Restaurant Data Analysis

## 📌 Overview
This project analyzes restaurant data to uncover insights related to cuisines, pricing, ratings, and online delivery trends.

The objective is to extract meaningful patterns that can help businesses make data-driven decisions.

---

## 🎯 Objectives
- Identify top cuisines
- Analyze city-wise restaurant distribution
- Study price range trends
- Understand online delivery patterns
- Explore restaurant ratings
- Detect geographic clusters
- Analyze restaurant chains

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import folium

sns.set(style="whitegrid")
```

## 📂 Dataset Description

The dataset contains information about restaurants including:
- Restaurant Name
- City
- Cuisines
- Ratings
- Price Range
- Online Delivery Availability

```python
df = pd.read_csv("/content/Dataset .csv")
df.head()
```

## 🧹 Data Cleaning

Steps performed:
- Handling missing values
- Removing duplicates
- Preparing data for analysis

```python
df.isnull().sum()
df.drop_duplicates(inplace=True)
```

## 🍜 Top Cuisines Analysis

### Objective
To identify the most popular cuisines.

```python
cuisines = df['Cuisines'].dropna().str.split(', ').explode()
top_cuisines = cuisines.value_counts().head(10)

top_cuisines.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Count")
plt.show()
```

### 📊 Insights
- Popular cuisines dominate customer preference
- Multi-cuisine offerings are very common

## 🏙️ City-wise Restaurant Distribution

### Objective
To analyze which cities have the highest number of restaurants.

```python
city_counts = df['City'].value_counts().head(10)

city_counts.plot(kind='bar', figsize=(10,5))
plt.title("Top Cities by Restaurant Count")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.show()
```

### 📊 Insights
- Metro cities have higher restaurant density
- Smaller cities have fewer restaurant options

## 💰 Price Range Distribution

### Objective
To understand affordability patterns.

```python
df['Price range'].value_counts().sort_index().plot(kind='bar')
plt.title("Price Range Distribution")
plt.xlabel("Price Range")
plt.ylabel("Count")
plt.show()
```

### 📊 Insights
- Most restaurants fall into mid-price categories
- High-end restaurants are fewer in number

## 🛵 Online Delivery Insights

### Objective
To study delivery availability trends.

```python
df['Has Online delivery'].value_counts().plot(kind='bar')
plt.title("Online Delivery Availability")
plt.show()
```

### 📊 Insights
- Online delivery is common in urban areas
- Restaurants offering delivery have better reach

## ⭐ Ratings Analysis

### Objective
To analyze customer satisfaction.

```python
df['Aggregate rating'].plot(kind='hist', bins=20)
plt.title("Ratings Distribution")
plt.xlabel("Rating")
plt.show()
```

### 📊 Insights
- Most ratings lie between 3.5 and 4.5
- Extremely low ratings are rare

## 🗺️ Geographic Clustering

### Objective
To visualize restaurant locations.

```python
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=5)

for i in range(100):
    folium.Marker(
        location=[df.iloc[i]['Latitude'], df.iloc[i]['Longitude']]
    ).add_to(restaurant_map)

restaurant_map
```

### 📊 Insights
- Restaurants cluster in high-density urban areas
- Potential opportunities exist in less dense regions

## 🔗 Restaurant Chain Analysis

### Objective
To identify popular chains.

```python
df['Restaurant Name'].value_counts().head(10)
```

### 📊 Insights
- Some restaurant brands dominate across cities
- Independent restaurants still form the majority

- Mid-range pricing dominates  
- Online delivery improves reach
